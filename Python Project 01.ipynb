Two columns of the dataframe (`price` and `odometer`) contain some non-numeric characters which need to be removed. Then, all the values need to be converted to a numerical data type.

# Dealing with Incorrect Registration Year Data

As we can see in the previous step of analysis, registration year in the dataset ranges from 1000 to 9999.  
  
According to the data available online, the first car in the world was built just before the year 1890. Thus, we can assume that neither car could be registered before the year 1890.  
  
As it comes to the maximum, neither car could have the first registration date later than the date when it was put on sale.  
Therefore, the maximum value cannot be greater than 2016.

```python
cars_1890_2016 = autos[(autos['registration_year'] >= 1890) & (autos['registration_year'] <= 2016)]
cars_1890_2016
```

```python
# calculate the distribution of the values in the 'registration_year' column
cars_1890_2016['registration_year'].value_counts(normalize = True)
```

Now, the years belong to the range 1890-2016.  
Most cars were registered around the year 2000.

# Exploring Price by Brand

```python
# print top 20 brands with the biggest numbers of cars for sale
top20 = autos['brand'].value_counts().head(20)
top20
```

```python
# create an empty dictionary which will contain average prices of cars grouped by brand names
average_prices = {}
```

```python
# loop through brand names saved in the 'top20' variable, calculate the averages, and add those values to the dictionary
for brand in top20.keys():
    avg_price = autos[autos['brand'] == brand]['price'].mean().round(2)
    average_prices[brand] = avg_price
```

```python
# print the results saved in a dictionary
average_prices
```

```python
# sort the results by average prices in ascending order
dict(sorted(average_prices.items(), key=lambda item: item[1]))
```

Average prices of these 20 car brands range from about 2,350 USD to over 40,000 USD.  
The lowest average value corresponds to Renault, the highest to Citroen.
